## 1. Pré-processamento dos dados - Parte 1

Cada ano tem um arquivo próprio, salvo na pasta [arquivo](./arquivos/dados_brutos/).

Para garantir que as mesmas métricas de pré-processamento sejam aplicadas, a função [preprocess_data](./funcoes/eda/preprocess_data.py) é usada.

Em resumo, os passos da função são os seguintes:
* Carrega o arquivo .csv de acordo com o ano de entrada.
* Limpa o nome das colunas com a remoção de tags HTML.
* Remove colunas que possuem 50%, ou mais, de dados nulos/faltantes.
* Realiza a tradução do título das colunas de inglês para português.
* Remove colunas que contem as palavras chave. 
* Verifica a porcentagem de valores faltantes nas colunas restantes. Estas serão avaliadas em uma etapa posterior.
* Por fim, salva os arquivos pré-processados na pasta [pre_process](./arquivos/pre_process/).


Justificativa de porque as remoções foram feitas:

1. Remover colunas do DataFrame que contêm as palavras-chave 'EUA', 'Por que', 'anterior', 'anteriores', 'país', 'UTC', 'raça' e 'ID' pode ser importante por alguns motivos:

* Relevância para a análise: Essas palavras-chave podem indicar colunas que  não são relevantes para a análise específica que está sendo realizada. Ao remover essas colunas, é possível focar nos dados mais pertinentes e reduzir o ruído nos resultados.

* Proteção de dados sensíveis: Algumas colunas podem conter informações sensíveis ou identificáveis, como informações de identificação pessoal (ID) ou dados relacionados a raça ou país. Remover essas colunas ajuda a proteger a privacidade e a conformidade com regulamentações de proteção de dados.

* Eliminação de variáveis irrelevantes: Palavras-chave como 'EUA', 'Por que' e 'anteriores' podem indicar perguntas ou respostas que não são relevantes para a análise em questão. A remoção dessas colunas pode simplificar o conjunto de dados e torná-lo mais focado nas variáveis de interesse.
